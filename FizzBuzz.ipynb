```python
def fizz_one():
    for val in xrange(1, 101):
        three_divides = not val % 3
        five_divides = not val % 5
        if three_divides and five_divides:
            print 'FizzBuzz'
        elif three_divides:
            print 'Fizz'
        elif five_divides:
            print 'Buzz'
        else:
            print val
```

```python
fizz_one()
```

```python
def fizz_two():
    for val in xrange(1, 101):
        string = ''
        if not val % 3:
            string += 'Fizz'
        if not val % 5:
            string += 'Buzz'
        print string
```

```python
fizz_two()
```

```python
def fizzbuzz():
    current = 0
    
    while current < 100:
        current += 1
        
        value = "{fizz}{buzz}".format(
            fizz="" if current % 3 else "Fizz",
            buzz="" if current % 5 else "Buzz")
        
        yield value if value else current

for shenanigans in fizzbuzz():
    print shenanigans
```
