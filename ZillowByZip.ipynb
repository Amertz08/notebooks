Generator that yields the string representation with the year and month of the last N months < present month

```python
def last_N_months(today, N):
    if today.month > N:
        for n in range(today.month - N, today.month):
                yield datetime.date(today.year, n, today.day).strftime('%Y-%m')
    else:
        for year in range(today.year - (N / 12), today.year + 1):
            for month in range(today.month + 12 - N, today.month):
                yield datetime.date(year, month, today.day).strftime('%Y-%m')
```

Returns the average price of homes based on zip code and month of the year

```python
def price_for_zip_by_month(zip_code, month):
    val = data[data.RegionName == zip_code][month].values
    if len(val) > 0:
        return val[0]
    else:
        return 0
```
